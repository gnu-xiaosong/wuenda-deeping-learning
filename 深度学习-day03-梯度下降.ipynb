# 梯度下降算法——Gradient descent

* function: to minimize other function as well—-最小化函数
* 代价函数$J(\theta_0, \theta_1....)$ 注意由于特征x输入值已知，因此$J(\theta_0, \theta_1....)$是关于$\theta_0, \theta_1.....$模型假设函数$F$的参数
* 目的就是求解最小化代价函数的参数$\theta_0, \theta_1....$

### 推导

* Hava some function: $J(\theta_0, \theta_1)$

* Want: $minJ(\theta_0, \theta_1)$

1. Start with some $\theta_0, \theta_1$-——初始化参数,通常给参数都初始化为0 $\theta=0$

2.Keep changing $\theta_0, \theta_1$ to reduce $J(\theta_0,\theta_1)$ until we hopefully end up at a minimum——找到使$J(\theta_0, \theta_1)$ 最小的参数$\theta_0,\theta_1$

> **思路**：从参数$\theta$初始化开始一点一点的改变$\theta$的大小使 $J_{\theta}$的逐渐变小，直到找到$J_{\theta}$的最小值或者局部最小值为至
