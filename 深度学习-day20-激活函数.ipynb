# activation function-激活函数

* hidden layer(隐藏层)：推荐使用RuLu,也可以用tanh

* output layer(输出层)：推荐使用sigmod。主要用于二元分类中，其他建议不要用！！

* ReLu使用最广，其他会梯度下降到倒数为0时下降缓慢的情况

## three activation function（三种激活函数的优缺点）

```python
import numpy as np
from matplotlib import pyplot as plt
```

* sigmod function

```python
x = np.linspace(-10, 10, 40)

y = 1 / (1 + np.exp(-x))

plt.figure()
plt.title('sigmod')

#Get the current axes
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('axes', 0.5))


plt.xlabel('x')
plt.ylabel('y')
plt.text(0, 0.5, '0.5', )
plt.annotate('$(0, 1)$', (5, 0.5))


plt.plot(x, y)
plt.show()
```

* tanh function

```python
x = np.linspace(-10, 10, 40)

y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.figure()
plt.title('tanh')

#Get the current axes
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('axes', 0.5))


plt.xlabel('x')
plt.ylabel('y')
plt.text(0, 0.5, '0.5', )
plt.annotate('$(-1, 1)$', (5, 0.5))


plt.plot(x, y)
plt.show()
```

* ReLu function:

```python
x = np.linspace(-20, 20, 40)

y = []
for i in range(np.size(x)):
    if x[i] <= 0:
        y.append(0)
    else:
        y.append(x[i])

y = np.array(y)

plt.figure()
plt.title('ReLu')

#Get the current axes
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('axes', 0.5))


plt.xlabel('x')
plt.ylabel('y')
plt.text(0, 0.5, '0.5', )
plt.annotate('$y = x$', (10, 10))
plt.annotate('$y = 0$', (-10, 0))

plt.plot(x, y)
plt.show()
```
