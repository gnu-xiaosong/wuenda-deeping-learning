## 正则化解决过拟合

* add pulishment-添加惩罚项

> 在cost function 中加入惩罚项 使参数$\theta_{3}和\theta_{4}$尽可能的小

由于不知道哪一些特征不太重要（惩罚），所以给所有特征加上惩罚项，使过于拟合的模型趋向于恰当的拟合模型。即是惩罚一些不太重要的特征。

* 加上惩罚项的正则化损失函数

$J(\theta) = \frac{1}{2m}[\sum_{i=1}^{m}(h_{\theta}(x^{(i)})-y^{(i)}) + \lambda\sum_{j=1}^{n}\theta_{j}^{2}]$

> $\lambda$-正则系数

  默认不惩罚$\theta_{0}$

* 高度正则化思想-避免过拟合
