### 总结

* when using the no-linear activation that the same of Logistic regression with neural network-使用非线性激活函数就和Logistic逻辑回归没有什么区别。neural networks hidden layer 就没有什么实际意义了，不管neurarl network layer 有多么多，最终结果还是特征的线性组合，也就是Logistic regression

* neural network activation function to choose: ReLu > tanh

* two classification(二元分类）：sigmod 概率输出

* about prediction value(关于预测类）：linear activation function （使用线性激活函数，比如房价预测类）
