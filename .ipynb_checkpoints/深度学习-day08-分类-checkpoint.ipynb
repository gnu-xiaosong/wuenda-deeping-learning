为了让$0<= h_{\theta}(x) <= 1$,我们引入了一个激活函数sigmod函数

```python
# sigmod function
import numpy as np
from matplotlib import pyplot as plt

x = 0.12 * np.arange(-100, 100, 0.01)
y = 1/(1 + np.exp(-x))

plt.figure()
plt.title('$sigmod function')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.scatter([0], [0.5], c='red')
plt.show()
```

## Cost Function 代价函数

讨论

* 最终代价函数表达式

$Cost(h_{\theta},y) =$ $$ \left\{
\begin{aligned}
-log(h_{\theta}(x)) -----if y = 1\\
-log(1-h_{\theta}(x)) ----- if y = 0
\end{aligned}
\right.$$

## 简化代价函数和梯度      ---simplified cost function and gradient descent                                   

损失函数公式-cost function

$J(\theta)=\frac{1}{m}\sum_{i=1}^{m}Cost(h_{\theta}(x),y^{(i)})$
       $=\frac{1}{m}[\sum_{i=1}^{m}y^{(i)} logh_{\theta}(x^{(i)})+(1-y^{(i)})log(1-h_{\theta}(x^{(i)})]$      
To fit parameters:

$minJ(\theta)$

To make a prediction given new:


Output  $h_{\theta} = \frac{1}{1+e^{- \theta^{T} X}}$



## Gradient Descent（梯度下降求解parameers) ---parameters

* cost function：

$J(\theta)=\frac{1}{m}\sum_{i=1}^{m}Cost(h_{\theta}(x),y^{(i)})$
       $=\frac{1}{m}[\sum_{i=1}^{m}y^{(i)} logh_{\theta}(x^{(i)})+(1-y^{(i)})log(1-h_{\theta}(x^{(i)})]$     


* wants to: **$minJ(\theta)$**

* diffenrent with Linear regression GradientDescent 与线性回归方程的不同
