* example one: regression---Housing price prediction

* image:

````python
#绘制图形
x = np.linspace(-5, 5, 50)
y = np.sin(x) + np.random.rand()

plt.figure
plt.title('Housing price prediction')
plt.xlabel('Price')
plt.ylabel('size in feet')
plt.scatter(x, y,label='数据点')

#模型训练
if  curve_matching(x, y):
    #训练成功-预测结果
    predic_y = predict(x)
    #绘制曲线
    plt.plot(x, predic_y, 'r-', label='拟合曲线')
    plt.annotate('最小二乘法拟合', (-2, 0), (0, 1))
    plt.legend() #显示图例

plt.show()

size = float(input('size:'))
#结果预测
pre = 'predict price:{0}'.format(predict(size))
print(pre)
````

````python
# curve matching(曲线拟合函数)
def curve_matching(x, y):
    '''
      利用最小二乘法拟合数据
      :param y: y for ndarray
             x: x for ndarrary
      :return expreess for curve matching
    '''
    param = np.polyfit(x, y, deg=10) #拟合阶数=5的多项式  返回降幂排序的多项式系数
    #保存模型数据
    np.savetxt('model.txt', param)
    return True

#模型预测
def predict(x):
    '''
    预测多形式模型---模型降幂构造
      :param x: var_able
      :return predict result
    '''
    #读取模型参数
    param = np.loadtxt('model.txt')
    #模型构造返回模型对象-方法一
    p = np.poly1d(param)
    return p(x)
````

* example two: calssifaction--Breast cancer(malignant, benign)

   one feature:

````python
y = np.array([0, 0, 0, 1, 0, 1, 0 , 1, 1,1,1])
x = np.linspace(1, 11, 11)


def color(y):
    c =[]
    for i in y:
         if i == 0:
            c.append('#1f77b4')
         else:
            c.append('#ff7f0e')
    
    return c

    
print(x)
plt.figure()
plt.title('Breast cancer(malignant, benign)')
plt.xlim(0, 12)
plt.ylim(0, 1.5)
plt.annotate('良性',(1,0), (0.8, 0.2))
plt.annotate('恶性',(4,1))
plt.xlabel('Tumor Size')
plt.ylabel('Malignant')
plt.scatter(x, y,c = color(y))

plt.show()
````

two feature:

````python
x = np.linspace(0, 100, 99)
y = -1 * x + 100


def different(y1):
    c = []
    for index in range(len(y1)):
        if y[index] - y1[index] < 0: 
            c.append('#1f77b4')
        elif y[index] - y1[index] == 0:
            c.append('#21d789')
        else:
            c.append('#ff7f0e')  
    return c


breast_y = np.random.randint(0, 100, 99)

plt.figure()

plt.title('Age-Tumor Size')
plt.xlim(0, 100)
plt.ylim(0, 100)

plt.xlabel('Tumor Size')
plt.ylabel('Age')
plt.text(20,20,'良性', fontsize=20)
plt.text(80,80,'恶性',fontsize=20)

plt.scatter(x, breast_y, c = different(breast_y), label='数据点')
plt.legend()
plt.plot(x, y)

plt.show()
````

> this preblem is calssifaction  这是一个分类问题 核心是怎样求解该分类曲线

## 无监督学习（Unsupervised Learning)

#### for example --- dustering 聚类算法

````python
x = np.random.randint(0, 100, size = 50)
y = np.random.randint(0, 100, size = 50)

plt.figure()

plt.title('for example')

plt.xlabel('x')
plt.ylabel('y')

plt.scatter(x, y, c='#2dc2d9')

plt.show()
````

* 鸡尾酒会算法--从混合的音频中分离出不同频率的音频

* 核心算法-Octave实现

```
[W,s,v] = svd(repmat(sum(x.*x, 1),size(x,1),1).*x)*x');
```

* 参数介绍：

svd -- 奇异值分解----python对应：np.svd(ndarray)
