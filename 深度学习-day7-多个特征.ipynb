
# 多个特征（Multiple feature）

## Multivariable Linear regression 多元线性回归

* Multiplevariables 多元特征

$X = [ x_{1},x_{2},x_{3}...x_{n}]$

* Hypothesis 假设函数

$h_{\theta}(x) = \Theta^{T} X = \theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2}+...+ \theta_{n}x_{n}$

> $x_{0}^{i} = 1$

* Paraeters 参数

$\Theta = [ \theta_{0}, \theta_{1},\theta_{2}...\theta_{n}]$

### Gradient Descent for Multiple Variables 多元梯度下降

* Cost Function 损失函数

$J(\theta_0,\theta_1...,\theta_{n})=\frac{1}{2m}\sum_{i=1}^{m}(h_\theta({x^{(i)})} - y^{x(i)})^2$ 

* $\Theta$ update

Repeat{

$\theta_{j} = \theta_{j} - \alpha\frac{1}{m}\sum_{i=1}^{m}(h_\theta({x^{(i)})} - y^{x(i)})x_{j}^{(i)}$

}

推导过程：

> goal: 将特征值缩放到相近的范围内，利于快速找到梯度最小值，迭代次数越少

## Feature Scaling I 特征缩放I----addition Max for feature （除以feature的最大值)

* Idea: make sure feature are on a similar scale.确保特征值在相近的范围内。then gradient descents can converge more quickly.这样的梯度下降法就能更快的收敛。

* Eg：
    $x_{1} = size(0-2000 feets_{2})$
