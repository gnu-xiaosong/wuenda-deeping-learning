# 随机初始化-parameters to randomly

> 随即初始化参数由于w具有完全性，故所以不能直接初始化参数为0.如果初始化为0，那么隐藏层就没有意义了。而b不一样不具有完全对称性。可以初始化为0

* 还有w也不能初始化太大的值，否则在激活函数中计算偏导数会接近0不易于处理

## 初始化实例

```python
import numpy as np

t = 0.01 # 为100时 由于激活函数会迫近极值0或1
w1 = np.random.randn(3,2) * t
print('w初始化为=',w1)

b1 = np.zeros((3, 1))
print('b初始化为=',b1)
```

# 总结

* $w$由于具有对称性故不能初始化为0，由于激活函数的自变量不能太大，会导致数据接近极值状态也不能初始化为太大的数。推荐初始化为[-1, 1]


```python
t = 0.01 # 为100时 由于激活函数会迫近极值0或1
w1 = np.random.randn(3,2) * t
print('w初始化为=',w1)
```

* $b$由于不具备完全对称性，故可以初始化为0

```python
b1 = np.zeros((3, 1))
print('b初始化为=',b1)
```
